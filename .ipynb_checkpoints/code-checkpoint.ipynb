## Data cleaning

```python
opened_file = open('AppleStore.csv',encoding='utf8')
opened_file1 = open('googleplaystore.csv',encoding='utf8')
from csv import reader 
read_file= reader(opened_file)
read_file1 = reader(opened_file1)
ios_data = list(read_file)
android_data = list(read_file1)
```

```python
def explore_data(dataset, start, end, rnc=False):
    dataset_slice = dataset[start:end]
    for i in dataset_slice:
        print(i)
        print('\n')
        
    if rnc:
        print('Number of rows: ', len(dataset))
        print('Number of columns: ', len(dataset[0]))
```

```python
duplicate_android=[]
unique_android=[]
for i in android_data[1:]:
    name = i[0]
    if name in unique_android:
        duplicate_android.append(name)
    else:
        unique_android.append(name)
print('Number of duplicate for Android: ', len(duplicate_android))
print('Example of duplicate for android: ', duplicate_android[:15])
```

```python
len(unique_android)
```

**Now we can see there will be 9659 rows of unique data**

```python
reviews_max_android={}
for i in android_data[1:]:
    name = i[0]
    n_reviews = float(i[3])
    if name in reviews_max_android and reviews_max_android[name]<n_reviews:
        reviews_max_android[name] = n_reviews
    elif name not in reviews_max_android:
        reviews_max_android[name] = n_reviews
        
```

```python
print("Actual unique data: ", len(reviews_max_android))
```

**Use dictionary to loop all the unique data. When the loop enconters a duplicated app name with a higher number of review, it will automatically update the number of reviews.**

```python
android_clean=[]
already_added=[]
for i in android_data[1:]:
    name = i[0]
    n_reviews = float(i[3])
    if n_reviews == reviews_max_android[name] and name not in already_added:
        android_clean.append(i)
        already_added.append(name)
```

```python
len(android_clean)
```

**Creating a list of list of clean android data. Using loop to compare the number of reviews with the number in reviews_max_android dictionary. When it is unique, it goes to android_clean and also marked down in the list of already_added.**

```python
explore_data(android_clean,0,5,True)
```

```python
def eng_check(string):
    non_ASCII=0
    for i in string:
        if ord(i)>127:
            non_ASCII +=1
    if non_ASCII<=3:
        return True
    else:
        return False
```

**A eng_check funtion to see if an app is english or non_english**
<br>-In order to not exclude the apps with emoji or some special characters, we allow apps to have up to 3 non_ASCII characters, which will be ord()>127<br/>

```python
ios_english=[]
android_english=[]
for i in ios_data[1:]:
    name = i[1]
    if eng_check(name):
        ios_english.append(i)

for i in android_clean:
    name = i[0]
    if eng_check(name):
        android_english.append(i)
    
```

## Data analysis

**Assume our company's goal is too:**
<br>-Build a minimal Android version of the app, and add it to Google Play.</br>
<br>-If the app has a good response from users, we develop it further.</br>
<br>-If the app is profitable after six months, we build an iOS version of the app and add it to the App Store.</br>

```python
def freq_table(dataset,index):
    frequency_count={}
    total=0
    for i in dataset:
        data = i[index]
        total+=1
        if data in frequency_count:
            frequency_count[data] +=1 
        else:
            frequency_count[data] =1
    
    frequency_percentage={}
    for i in frequency_count:
        percentage = (frequency_count[i]/total)*100
        frequency_percentage[i]=percentage
    return frequency_percentage
```

```python
print(freq_table(ios_english_free,11))
```

```python
def display_table(dataset,index):
    table = freq_table(dataset,index)
    table_display=[]
    for i in table:
        i_tuple = (table[i],i)
        table_display.append(i_tuple)
    table_sorted = sorted(table_display,reverse= True)
    for i in table_sorted:
        print(i[1],':',i[0])
```

```python
display_table(ios_english_free,11)
```

## ios_app analysis

```python
genre_ios = freq_table(ios_english_free,11)
for genre in genre_ios:
    total=0
    len_genre=0
    for app in ios_english_free:
        genre_app = app[11]
        if genre_app == genre:
            count_total = float(app[5])
            total+=count_total
            len_genre+=1
    avg_count_total = total/len_genre
    print(genre, ':', round(avg_count_total))
```

**As we can see, Social Networking, Navigation, Reference and Weather have a relatively higher number of avg rating. The next step, we want to see if the data is skewed in a specific genre, make sure the mean is not influenced too much by those most popular apps**

## Android_app analysis

```python
categories_android = freq_table(android_english_free,1)
types_review={}
for category in categories_android:
    total=0
    len_category=0
    for app in android_english_free:
        category_app = app[1]
        if category_app == category:
            count_total = app[5]
            count_total = count_total.replace(',','')
            count_total = count_total.replace('+','')
            total+=float(count_total)
            len_category+=1
    avg_total = total/len_category
    if category not in types_review:
        types_review[category]=round(avg_total) 

dict(sorted(types_review.items(), key=lambda item: item[1],reverse=True))
```

```python
for app in android_english_free:
    if app[1] == 'COMMUNICATION' and (app[5] == '1,000,000,000+'
                                            or app[5] == '500,000,000+'
                                            or app[5] == '100,000,000+'):
        print(app[0], ':', app[5])
```
